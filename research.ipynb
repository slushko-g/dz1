Перед тем как удалять NaN в RainToday хорошо бы понять как влияют остальные признаки на RainTomorrow. Попробуем убрать все строки в которых RainToday не NaN и посмотреть корреляцию признаков с RainTommorow. Поэтому план далее выглядит так:

Можно выбрать OHE для направления ветра и месяцев, либо сделать их циклическими через sin и cos. Предпочтительнее второй вариант, так как он не сильно увеличит количество признаков и при этом сохранит необходимую информацию.
Для дерева можно будет сохранить месяц (т.е. пробуем линейную регрессию с энкодингом и дерево и без энкодинга)

После энкодинга, имеет смысл посмотреть матрицу корреляции и решить, можно ли выкинуть те 2% записей, где RainToday = NaN

Переведем направление ветра в радианы. Посмотрим уникальные значения для направлений ветра

```python
df_cycle = df.copy()

print(df_cycle['WindGustDir'].unique())
print(df_cycle['WindDir3pm'].unique())
print(df_cycle['WindDir9am'].unique())
```

```python
wind_to_deg = {
    'N': 0.0, 'NNE': 22.5, 'NE': 45.0, 'ENE': 67.5,
    'E': 90.0, 'ESE': 112.5, 'SE': 135.0, 'SSE': 157.5,
    'S': 180.0, 'SSW': 202.5, 'SW': 225.0, 'WSW': 247.5,
    'W': 270.0, 'WNW': 292.5, 'NW': 315.0, 'NNW': 337.5
}

wind_cols = ["WindGustDir", "WindDir9am", "WindDir3pm"]

for col in wind_cols:
    deg = df_cycle[col].map(wind_to_deg)          
    rad = np.deg2rad(deg)                   
    df_cycle[col + "_sin"] = np.sin(rad)
    df_cycle[col + "_cos"] = np.cos(rad)

df_cycle = df_cycle.drop(columns=wind_cols)
```

Наверное, имеет смысл извлечь из даты только месяц, а остальное отбросить. "зациклизуем" значение, чтобы осталась информация что декабрь близко к январю. Создадим два новых столбца - month_sin и month_cos

```python
df_cycle['Date'] = pd.to_datetime(df_cycle['Date']).dt.month
df_cycle['month_sin'] = np.sin(2 * np.pi * (df_cycle['Date'] - 1) / 12)
df_cycle['month_cos'] = np.cos(2 * np.pi * (df_cycle['Date'] - 1) / 12)
df_cycle = df_cycle.drop('Date', axis=1)
```

# Пробуем собрать logistic regression baseline

```python
target = "RainTomorrow"
cat_features = ["Location", "RainToday"]
num_features = [c for c in df_cycle.columns if c not in cat_features + [target]]
```

```python
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression

X = df_cycle.drop(columns=[target])
y = df_cycle[target].astype(int)

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median", add_indicator=True)),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_features),
        ("cat", categorical_transformer, cat_features),
    ],
    remainder="drop"
)

clf = Pipeline(steps=[
    ("preprocess", preprocess),
    ("model", LogisticRegression(max_iter=2000, class_weight="balanced", random_state=102))
])

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=102)

scoring = {
    "roc_auc": "roc_auc",
    "pr_auc": "average_precision",
    "f1": "f1",
    "recall": "recall",
    "precision": "precision"
}

res = cross_validate(clf, X, y, cv=cv, scoring=scoring, n_jobs=-1, return_train_score=True)

for k in scoring:
    mean = res[f"test_{k}"].mean()
    std = res[f"test_{k}"].std()
    print(f"{k}: {mean:.3f} ± {std:.3f}")
```

Попробуем подобрать гиперпараметры для логистической регрессии

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

C_grid = [10**p for p in [-3, -2.5, -2, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 2]]

logreg_pipe = Pipeline(steps=[
    ("preprocess", preprocess),
    ("model", LogisticRegression(max_iter=3000, solver="saga", random_state=102))
])

param_grid = [
    {"model__penalty": ["l2"], "model__C": C_grid, "model__class_weight": [None, "balanced"]},
    {"model__penalty": ["l1"], "model__C": C_grid, "model__class_weight": [None, "balanced"]},
    {"model__penalty": ["elasticnet"], "model__C": C_grid, "model__l1_ratio": [0.2, 0.5, 0.8], "model__class_weight": [None, "balanced"]},
]

gs = GridSearchCV(
    estimator=logreg_pipe,
    param_grid=param_grid,
    scoring="average_precision",
    cv=cv,
    n_jobs=-1,
    refit=True,
    return_train_score=True
)

gs.fit(X, y)
print(f"Best PR-AUC: {gs.best_score_:.3f}")
print("Best params:", gs.best_params_)
```

# Дерево baseline

```python
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier

target = "RainTomorrow"
cat_features = ["Location", "RainToday"]
num_features = [c for c in df_cycle.columns if c not in cat_features + [target]]

X = df_cycle.drop(columns=[target])
y = df_cycle[target].astype(int)

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median", add_indicator=True))
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_features),
        ("cat", categorical_transformer, cat_features),
    ],
    remainder="drop"
)

tree = Pipeline(steps=[
    ("preprocess", preprocess),
    ("model", DecisionTreeClassifier(random_state=102, class_weight="balanced"))
])

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=102)

scoring = {
    "roc_auc": "roc_auc",
    "pr_auc": "average_precision",
    "f1": "f1",
    "recall": "recall",
    "precision": "precision"
}

res = cross_validate(tree, X, y, cv=cv, scoring=scoring, n_jobs=-1, return_train_score=True)

for k in scoring:
    test_mean = res[f"test_{k}"].mean()
    test_std = res[f"test_{k}"].std()
    train_mean = res[f"train_{k}"].mean()
    print(f"{k}: test={test_mean:.3f}±{test_std:.3f} train={train_mean:.3f}")
```

Видно, что дерево переобучилось, поробуем с помощью grid search подобрать гиперпараметры

```python
settings = [
    {"max_depth": 3, "min_samples_leaf": 200, "min_samples_split": 400, "ccp_alpha": 0.0},
    {"max_depth": 5, "min_samples_leaf": 100, "min_samples_split": 200, "ccp_alpha": 0.0},
    {"max_depth": 7, "min_samples_leaf": 50, "min_samples_split": 100, "ccp_alpha": 0.0},
    {"max_depth": 5, "min_samples_leaf": 100, "min_samples_split": 200, "ccp_alpha": 1e-3},
    {"max_depth": None, "min_samples_leaf": 100, "min_samples_split": 200, "ccp_alpha": 1e-3},
]

for s in settings:
    model = Pipeline(steps=[
        ("preprocess", preprocess),
        ("model", DecisionTreeClassifier(random_state=102, class_weight="balanced", **s))
    ])
    res = cross_validate(model, X, y, cv=cv, scoring=scoring, n_jobs=-1, return_train_score=True)
    test_pr_mean = res["test_pr_auc"].mean()
    train_pr_mean = res["train_pr_auc"].mean()
    test_roc_mean = res["test_roc_auc"].mean()
    print(
        f"max_depth={s['max_depth']}, min_leaf={s['min_samples_leaf']}, min_split={s['min_samples_split']}, ccp_alpha={s['ccp_alpha']}: "
        f"PR-AUC test={test_pr_mean:.3f}, PR-AUC train={train_pr_mean:.3f}, ROC-AUC test={test_roc_mean:.3f}"
    )
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier

tree_pipe = Pipeline(steps=[
    ("preprocess", preprocess),
    ("model", DecisionTreeClassifier(random_state=102, class_weight="balanced"))
])

param_grid = {
    "model__max_depth": [6, 7, 8, 9, 10],
    "model__min_samples_leaf": [20, 30, 50, 80],
    "model__min_samples_split": [50, 100, 200],
    "model__ccp_alpha": [0.0, 1e-5, 1e-4, 5e-4, 1e-3],
    "model__max_features": [None, "sqrt", "log2", 0.5],
}

gs = GridSearchCV(
    estimator=tree_pipe,
    param_grid=param_grid,
    scoring="average_precision",
    cv=cv,
    n_jobs=-1,
    refit=True,
    return_train_score=True
)

gs.fit(X, y)
print(f"Best PR-AUC: {gs.best_score_:.3f}")
print("Best params:", gs.best_params_)
```
